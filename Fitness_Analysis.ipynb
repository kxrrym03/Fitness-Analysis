## Visualizations

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='activity_type')
plt.title('Distribution of Activity Types')
plt.xticks(rotation=270)
plt.xlabel('Activity Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='gender')
plt.title('Distribution of Gender')
plt.xticks(rotation=270)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['timestamp'] = df['timestamp'].dt.date
df.head()
```

```python
df_sorted = df.sort_values(by='timestamp')
df_sorted['steps_rolling_avg'] = df_sorted['steps'].rolling(window=7).mean()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_sorted, x='timestamp', y='steps', label='Daily Steps', alpha=0.6)
sns.lineplot(data=df_sorted, x='timestamp', y='steps_rolling_avg', label='7-Day Rolling Average')

plt.title('Steps over Time with Rolling Average')
plt.xlabel('Timestamp')
plt.ylabel('Steps')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
avg_calories_burned_by_activity = df.groupby('activity_type')['calories_burned'].mean().reset_index()
print("Average Calories Burned by Activity Type:")
print(avg_calories_burned_by_activity)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_calories_burned_by_activity, x='activity_type', y='calories_burned')
plt.title('Average Calories Burned by Activity Type')
plt.xlabel('Activity Type')
plt.ylabel('Average Calories Burned')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
average_weight_by_gender = df.groupby('gender')['weight_kg'].mean().reset_index()
print("Average Weight (kg) by Gender:")
print(average_weight_by_gender)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=average_weight_by_gender, x='gender', y='weight_kg', palette='viridis')
plt.title('Average Weight (kg) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Weight (kg)')
plt.tight_layout()
plt.show()
```

```python
avg_duration_by_activity = df.groupby('activity_type')['duration_min'].mean().reset_index()
print("Average Duration (minutes) by Activity Type:")
print(avg_duration_by_activity)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_duration_by_activity, x='activity_type', y='duration_min', palette='muted')
plt.title('Average Duration (minutes) by Activity Type')
plt.xlabel('Activity Type')
plt.ylabel('Average Duration (minutes)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
